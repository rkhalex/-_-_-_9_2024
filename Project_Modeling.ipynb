```python
# Function to plot countplots in a grid
def plot_countplots(data, title, n_cols, n_rows):
    # n_rows = (len(data.columns) + 1) // n_cols
    fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 5 * n_rows))
    fig.suptitle(title, fontsize=16)
    
    for i, column in enumerate(data.columns):
        ax = axes[i // n_cols, i % n_cols]
        sns.countplot(data=data, x=column, ax=ax, color='#ff7847')
        ax.set_title(f'Countplot of {column}')
        for label in ax.get_xticklabels():
            label.set_rotation(45)
    
    # Hide any empty subplots
    if len(data.columns) % n_cols != 0:
        for j in range(len(data.columns), n_cols * n_rows):
            fig.delaxes(axes[j // n_cols, j % n_cols])
    
    plt.tight_layout(rect=[0, 0, 1, 0.97])
    plt.show()
    
# Function to plot countplots in a grid
def plot_distplots(data, title, n_cols, n_rows):
    # n_rows = (len(data.columns) + 1) // n_cols
    fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 5 * n_rows))
    fig.suptitle(title, fontsize=16)
    
    for i, column in enumerate(data.columns):
        ax = axes[i // n_cols, i % n_cols]
        sns.histplot(data=data, x=column, ax=ax, color='#ff7847')
        ax.set_title(f'Countplot of {column}')
        for label in ax.get_xticklabels():
            label.set_rotation(45)
    
    # Hide any empty subplots
    if len(data.columns) % n_cols != 0:
        for j in range(len(data.columns), n_cols * n_rows):
            fig.delaxes(axes[j // n_cols, j % n_cols])
    
    plt.tight_layout(rect=[0, 0, 1, 0.97])
    plt.show()
    
    
```

## Cross-Validation

```python
from sklearn.model_selection import cross_val_score, StratifiedKFold
```

```python
# Cross-validation setup
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Perform cross-validation and calculate ROC AUC for each model
def perform_cv(model, X, y):
    scores = cross_val_score(model, X, y, cv=cv, scoring='roc_auc')
    return scores

# Logistic Regression
roc_auc_logreg_email = perform_cv(logreg_email, X, y_email)
roc_auc_logreg_phone = perform_cv(logreg_phone, X, y_phone)

# Random Forest
roc_auc_rf_email = perform_cv(rf_email, X, y_email)    

# Gradient Boosting
roc_auc_gb_email = perform_cv(gb_email, X, y_email)
roc_auc_gb_phone = perform_cv(gb_phone, X, y_phone)

# Print results
print("Logistic Regression - Email Response ROC AUC Scores:", roc_auc_logreg_email)
print("Logistic Regression - Email Response Mean ROC AUC:", np.mean(roc_auc_logreg_email))
print("Logistic Regression - Phone Response ROC AUC Scores:", roc_auc_logreg_phone)
print("Logistic Regression - Phone Response Mean ROC AUC:", np.mean(roc_auc_logreg_phone))

print("\nRandom Forest - Email Response ROC AUC Scores:", roc_auc_rf_email)
print("Random Forest - Email Response Mean ROC AUC:", np.mean(roc_auc_rf_email))
print("Random Forest - Phone Response ROC AUC Scores:", roc_auc_rf_phone)
print("Random Forest - Phone Response Mean ROC AUC:", np.mean(roc_auc_rf_phone))

print("\nGradient Boosting - Email Response ROC AUC Scores:", roc_auc_gb_email)
print("Gradient Boosting - Email Response Mean ROC AUC:", np.mean(roc_auc_gb_email))
print("Gradient Boosting - Phone Response ROC AUC Scores:", roc_auc_gb_phone)
print("Gradient Boosting - Phone Response Mean ROC AUC:", np.mean(roc_auc_gb_phone))
```

## Money predict

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


email_probabilities = rf_email.predict_proba(X_test_email)[:, 1]  # Probability of positive class for email
phone_probabilities = gb_phone.predict_proba(X_test_phone)[:, 1]  # Probability of positive class for phone


cost_per_email = 1  # Example cost
cost_per_phone = 10  # Example cost

# Calculate expected responses and costs for different thresholds
thresholds = np.linspace(0, 1, 1000)
expected_responses = []
costs = []

for threshold in thresholds:
    # Predict responses based on threshold
    predicted_email_responses = (email_probabilities >= threshold).astype(int)
    predicted_phone_responses = (phone_probabilities >= threshold).astype(int)
    
    # Calculate expected number of responses
    expected_email_response_count = np.sum(predicted_email_responses)
    expected_phone_response_count = np.sum(predicted_phone_responses)
    total_expected_responses = expected_email_response_count + expected_phone_response_count
    
    # Calculate cost
    total_cost = (expected_email_response_count * cost_per_email) + (expected_phone_response_count * cost_per_phone)
    
    if total_cost > 5000000:  # Stop if cost exceeds 500,000
            break

    expected_responses.append(total_expected_responses)
    costs.append(total_cost)

# Plot the graph
plt.figure(figsize=(12, 7))
plt.plot(costs, expected_responses, label='Expected Response vs. Cost')
plt.xlabel('Cost')
plt.ylabel('Expected Responses')
plt.title('Expected Response vs. Cost of Communication')
plt.legend()
plt.grid(True)
plt.show()
```
